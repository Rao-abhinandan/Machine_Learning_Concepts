Lets create a vibration data of shaft to understand statistcial concepts

```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Define the number of machines
num_machines = 100

# Generate data
machine_ids = [f"Machine_{i+1}" for i in range(num_machines)]
power_output = np.random.uniform(50, 200, num_machines)  # in kW
efficiency = np.random.uniform(70, 95, num_machines)     # in percentage
vibration_level = np.random.normal(0.5, 0.1, num_machines)  # in mm/s (mean=0.5, std=0.1)
failure_rate = np.random.exponential(0.1, num_machines)   # in failures per year

# Create a DataFrame
data = {
    "Machine ID": machine_ids,
    "Power Output (kW)": power_output,
    "Efficiency (%)": efficiency,
    "Vibration Level (mm/s)": vibration_level,
    "Failure Rate (failures/year)": failure_rate,
}

df = pd.DataFrame(data)

# Save the dataset to a CSV file
#df.to_csv("machine_performance_data.csv", index=False)
```

Lets compute mean, median and mode using python

*Mean (xÌ…)  = Sum of Values / Number of Values

*Median = [(n + 1)/2]th term  if n (total  no of observations) is odd 

 Median  = [(n/2)th term + {(n/2) + 1}th term] / 2  if n (total  no of observations) is even

*Mode = highest occuring term

```python
# del sum 
def calculate_std(data):
    mean = sum(data) / len(data)
    squared_diff = [(x - mean) ** 2 for x in data]  
    variance = sum(squared_diff) / len(data)
    std_dev = variance ** 0.5
    
    return std_dev


std_dev = calculate_std(data)

print(f"Standard Deviation: {std_dev}")
```

```python
import matplotlib.pyplot as plt


for values in df.columns[1:-1]:

    # mean, median and standard deviation
    mean = df[values].mean()
    median = df[values].median()
    std_dev = df[values].std()


    plt.figure(figsize=(8, 6))

    # Plot the histogram of the data
    plt.hist(df[values], bins=10, alpha=0.7, color='blue', label='Data')

    # Add lines for mean and median
    plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
    plt.axvline(median, color='green', linestyle='--', label=f'Median: {median:.2f}')
    plt.axvline(mean + std_dev, color='orange', linestyle=':', label=f'Mean + 1 Std Dev: {mean + std_dev:.2f}')
    plt.axvline(mean - std_dev, color='orange', linestyle=':', label=f'Mean - 1 Std Dev: {mean - std_dev:.2f}')
    # Adding labels and title
    plt.title('Histogram with Mean and Median')
    plt.xlabel(values)
    plt.ylabel('Frequency')

    # Show the legend
    plt.legend()

    # Show the plot
    plt.show()
```

```python
import seaborn as sns

for values in df.columns[1:-1]:

    # mean, median and standard deviation
    mean = df[values].mean()
    median = df[values].median()
    std_dev = df[values].std()

    # Plot KDE
    plt.figure(figsize=(8, 6))
    sns.kdeplot(df[values], shade=True, color='blue', label='KDE')

    # Add vertical lines for mean and median
    plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
    plt.axvline(median, color='green', linestyle='--', label=f'Median: {median:.2f}')
    plt.axvline(mean + std_dev, color='orange', linestyle=':', label=f'Mean + 1 Std Dev: {mean + std_dev:.2f}')
    plt.axvline(mean - std_dev, color='orange', linestyle=':', label=f'Mean - 1 Std Dev: {mean - std_dev:.2f}')    

    # Adding labels and title
    plt.title('KDE Plot with Mean and Median')
    plt.xlabel(values)
    plt.ylabel('Density')

    # Show the legend
    plt.legend()

    # Show the plot
    plt.show()
```
